# 8)	Write a python program to scrape the details of most downloaded articles from AI in last 90 days. https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details :
## i)	Paper Title
## ii)	Authors
## iii)	Published Date
## iv)	Paper URL
 


```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
url = requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
url
```

```python
soup=BeautifulSoup(url.content)
```

```python
articles = []

for i in soup.find_all('h2',class_='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
    articles.append(i.text.split())
    
articles_name = []
for ele in range(len(articles)):
    articles_name.append(' '.join(articles[ele]))

#articles_name
```

```python
authors = []

for i in soup.find_all('span',class_='sc-1w3fpd7-0 dnCnAO'):
    authors.append(i.text.split())
    
#print (authors)
```

```python
published_date = []

for i in soup.find_all('span',class_='sc-1thf9ly-2 dvggWt'):
    published_date.append(i.text.split())
    
#published_date
```

```python
paper_url = []

for i in soup.find_all('a', class_='sc-5smygv-0 fIXTHm'):
    paper_url.append(i.get('href').split())# =='https://www.sciencedirect.com/science/article/pii/'))
    
#print (paper_url)
```

```python
print (len (articles_name), len (authors), len (published_date), len (paper_url))
```

```python
AI_articles_df = pd.DataFrame({'Paper Title': articles_name, 'Authors' : authors, 'Published Date' : published_date, 'Paper URL' : paper_url})
AI_articles_df
```
