i)	Restaurant name
ii)	Cuisine
iii)Location
iv)	Ratings
v)	Image URL


```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings('ignore')
```

```python
url = requests.get("https://www.dineout.co.in/ahmedabad-restaurants")
url
```

```python
soup = BeautifulSoup(url.content)
```

```python

rest_name=[]

for i in soup.find_all('a',class_="restnt-name ellipsis"):
    rest_name.append(i.text)

#rest_name
```

```python
cuisine_name=[]

for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine_name.append(i.text.split('|')[1])
    
#cuisine_name
```

```python
location_name=[]

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location_name.append(i.text)

#location_name
```

```python
rating_name=[]
rating_name0=[]
rating_name1=[]
rating_name2=[]

for i in soup.find_all('div',class_="restnt-rating rating-10000"):
    rating_name0.append(i.text)

for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating_name1.append(i.text)
    
for i in soup.find_all('div',class_="restnt-rating rating-5"):
    rating_name2.append(i.text)
    
rating_name = rating_name0 + rating_name1 + rating_name2
#rating_name
```

```python
img_url=[]

for i in soup.find_all('img',class_="no-img"):
    img_url.append(i.get('data-src'))
    
#img_url
```

```python
print (len (rest_name), len(cuisine_name), len(location_name), len(rating_name), len(img_url))
```

```python
Restaurant_Detail_df = pd.DataFrame({'Restaurant name' : rest_name, 'Cuisine' : cuisine_name, 'Location' : location_name, 'Ratings' : rating_name, 'Image URL' : img_url})
```
