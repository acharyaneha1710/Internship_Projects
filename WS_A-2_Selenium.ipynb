```python
start=0
end=3
flipkart= range(start,end)

Brand=[]
Pdct_Des=[]
Pdct_Des1=[]
Pdct_Des2=[]
Price=[]
Disc=[]

for page in flipkart: 
    
    Brand_Name = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    
    Product1 = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    Product2 = driver.find_elements(By.XPATH,'//a[@class="IRpwTa _2-ICcC"]')
    
    Price_tag = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    Disc_off = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    next_btn= driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')

       
    if flipkart[page] < end-1:
        
        for i in Brand_Name[0:40]:
            Brand.append(i.text)
    
        for i in Product1[0:40]:
            Pdct_Des1.append(i.text)
            
        for i in Product2[0:40]:
            Pdct_Des2.append(i.text)
                    
        for i in Price_tag[0:40]:
            Price.append(i.text)
    
        for i in Disc_off[0:40]:
            Disc.append(i.text)
             
    elif flipkart[page] == end-1:
        
        for i in Brand_Name[0:20]:
            Brand.append(i.text)

        for i in Product1[0:20]:
            Pdct_Des1.append(i.text)
            
        for i in Product2[0:20]:
            Pdct_Des2.append(i.text)
                    
        for i in Price_tag[0:20]:
            Price.append(i.text)
    
        for i in Disc_off[0:20]:
            Disc.append(i.text)
            
            
    next_btn.click()
    time.sleep(1)
        
Pdct_Des= Pdct_Des1 + Pdct_Des2       

print(len(Brand), len(Pdct_Des), len(Price), len(Disc))
```

```python
FK_df=pd.DataFrame({'Brand Name':Brand, 'Product Description':Pdct_Des, 'Price':Price, 'Discount':Disc})
FK_df
```

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.azquotes.com/')
```

```python
Top_Quotes= driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
Top_Quotes.click()
```

```python
start=0
end=10


Quote=[]
Author=[]
Type_Quotes=[]

full_quote = driver.find_elements(By.XPATH,'//a[@class="title"]')
author_name = driver.find_elements(By.XPATH,'//div[@class="author"]')
quote_type = driver.find_elements(By.XPATH,'//div[@class="tags"]')

next_btn= driver.find_element(By.XPATH,'//div[@class="pager"]')

for i in range (start,end):
    
    for i in full_quote[0:100]:
        Quote.append(i.text)
        
    for i in author_name[0:100]:
        Author.append(i.text)
        
    for i in quote_type[0:100]:
        Type_Quotes.append(i.text)
        
    next_btn.click()
    time.sleep(2)
    
print(len(Quote), len(Author), len(Type_Quotes))
```

```python
author_df= pd.DataFrame({'Quote':Quote, 'Author Name':Author, 'Types of Quote':Type_Quotes})
author_df
```
