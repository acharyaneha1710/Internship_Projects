```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https:www.flipkart.com')
```

```python
close_loginpage = driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button')
close_loginpage.click()
```

```python
srch = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
srch.send_keys('oneplus nord')
srch.submit()
```

```python
brand,name,color = [],[],[]
for i in driver.find_elements(By.XPATH,'//div[@class = "_4rR01T"]'):
    brand.append(i.text.split()[0])
    name.append(i.text.split()[1:4])
    color.append(i.text.split()[-4:-2])
```

```python
ram,rom,cam,display,battery,price,url,pri = [],[],[],[],[],[],[],[]

for i in driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]'):
    ra = i.text.split('\n')[0]
    ram.append(ra.split('|')[0])
    rom.append(ra.split('|')[1])
    
    cam.append(i.text.split('\n')[2])
    display.append(i.text.split('\n')[1])
    battery.append(i.text.split('\n')[-2])

for i in driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]'): 
    price.append(i.text)
    
for i in driver.find_elements(By.XPATH,'//div[@class="_3utEwz"]'):
    pri.append(i.text)
        
price.append(pri)
    
for i in driver.find_elements(By.XPATH,"//a[@class = '_1fQZEK']"):
    url.append(i.get_attribute('href'))

len(cam)
```

```python
print (len(brand), len(name), len(color), len(ram), len(rom), len(cam), len(display), len(battery), len(price), len(url))
```

```python
oneplus_df=pd.DataFrame({'Brand Name':brand,'Smartphone name':name, 'Colour':color, 'RAM':ram, 'Storage(ROM)':rom, 'Primary Camera':cam, 'Display Size':display, 'Battery Capacity':battery, 'Price':price, 'Product URL':url})
oneplus_df.to_csv('oneplus_data.csv')
```
